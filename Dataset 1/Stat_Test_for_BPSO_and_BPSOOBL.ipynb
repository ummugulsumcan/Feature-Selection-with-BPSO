```python
import pandas as pd
import numpy as np
BPSO = pd.read_csv("Dataset1_BPSO.txt", index_col=False)
BPSO_OBL = pd.read_csv("Dataset1_BPSO+OBL.txt", index_col=False)
SFS = pd.read_csv("Dataset1_SFS.txt", index_col=False)
Results = pd.DataFrame()
Results2= pd.DataFrame()
```

```python
Results['BPSO']= BPSO['Train Score']
Results['BPSO_OBL']=BPSO_OBL['Train Score']
Results['SFS'] = SFS['Train Score']
```

```python
Results2['BPSO_2']= BPSO['Test Score']
Results2['BPSO_OBL_2']=BPSO_OBL['Test Score']
Results2['SFS_2'] = SFS['Test Score']
```

```python
ResultsTrain = pd.melt(Results,var_name='Type', value_name='Train_Score')
ResultsTrain = ResultsTrain.reset_index(drop=True)

Results2Test = pd.melt(Results2,var_name='Type', value_name='Test_Score')
Results2Test = Results2Test.reset_index(drop=True)
```

```python
ResultsTrain
```

```python
Results2Test
```

```python
ResultsTrain['Type']= ResultsTrain['Type'].astype('category')
Results2Test['Type']= Results2Test['Type'].astype('category')
```

```python
import matplotlib.pyplot as plt
import seaborn
%matplotlib inline 
```

```python
#Train Score
import statsmodels.formula.api as smf
import statsmodels.stats.multicomp as multi 

model1= smf.ols('Train_Score ~ C(Type)', data=ResultsTrain)
results= model1.fit()
print(results.summary())
```

```python
mc1 = multi.MultiComparison(ResultsTrain['Train_Score'], ResultsTrain['Type'])
res1 = mc1.tukeyhsd()
print(res1.summary())
```

```python
#Test Scores
import statsmodels.formula.api as smf
import statsmodels.stats.multicomp as multi 

model1= smf.ols('Test_Score ~ C(Type)', data=Results2Test)
results= model1.fit()
print(results.summary())
```

```python
mc1 = multi.MultiComparison(Results2Test['Test_Score'], Results2Test['Type'])
res1 = mc1.tukeyhsd()
print(res1.summary())
```

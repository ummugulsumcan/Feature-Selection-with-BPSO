```python
import pandas as pd
import numpy as np
BPSO = pd.read_csv("Dataset2_BPSO.txt", index_col=False)
BPSO_OBL = pd.read_csv("Dataset2_BPSO+OBL.txt", index_col=False)
SFS = pd.read_csv("Dataset2_SFS.txt", index_col=False)
Results = pd.DataFrame()
Results2= pd.DataFrame()
```

```python
Results['BPSO']= BPSO['Train Score']
Results['BPSO_OBL']=BPSO_OBL['Train Score']
Results['SFS'] = SFS['Train Score']
```

```python
Results2['BPSO_2']= BPSO['Test Score']
Results2['BPSO_OBL_2']=BPSO_OBL['Test Score']
Results2['SFS_2'] = SFS['Test Score']
```

```python
Results = pd.melt(Results,var_name='Type', value_name='Train_Score')
Results = Results.reset_index(drop=True)

Results2 = pd.melt(Results2,var_name='Type', value_name='Test_Score')
Results2 = Results2.reset_index(drop=True)
```

```python
Results['Type']= Results['Type'].astype('category')
Results2['Type']= Results2['Type'].astype('category')
```

```python
import matplotlib.pyplot as plt
import seaborn
%matplotlib inline 
```

```python
#Train Score
import statsmodels.formula.api as smf
import statsmodels.stats.multicomp as multi 

model1= smf.ols('Train_Score ~ C(Type)', data=Results)
results= model1.fit()
print(results.summary())
```

```python
mc1 = multi.MultiComparison(Results['Train_Score'], Results['Type'])
res1 = mc1.tukeyhsd()
print(res1.summary())
```

```python
#Test Scores
import statsmodels.formula.api as smf
import statsmodels.stats.multicomp as multi 

model1= smf.ols('Test_Score ~ C(Type)', data=Results2)
results= model1.fit()
print(results.summary())
```

```python
mc1 = multi.MultiComparison(Results2['Test_Score'], Results2['Type'])
res1 = mc1.tukeyhsd()
print(res1.summary())
```
